### Заполнение пропусков в locality_name

```python
# Проверим данные в столбце locality_name:
data['locality_name'].unique()
```

Есть много повторов, исправим повторы в названиях населенных пунктов, создав соответствующую функцию:

```python
data['locality_name'] = data['locality_name'].fillna('нет данных')
```

```python
def locality_name(location):
    if 'петер'in location or 'питер' in location:
        return 'Сантк-Петербург'
    if 'Мурин' in location:
        return 'Мурино'
    if 'Кудров' in location:
        return 'Кудрово'
    if 'Шушар' in location:
        return 'Шушары'
    if 'Всеволож' in location:
        return 'Всеволожск'
    if 'Паргол' in location:
        return 'Парголово'
    if 'Колпин' in location:
        return 'Колпино'
    if 'Пушкин' in location:
        return 'Пушкин'
    if 'Гатчин' in location:
        return 'Гатчина'
    if 'Выборг' in location:
        return 'Выборг'
    return location
data['locality_name'] = data['locality_name'].apply(locality_name)
```

## Расчёты и добавление результатов в таблицу

Рассчитаем для исследований следующие параметры и добавим в таблицу данных:
* цена квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры (варианты — первый, последний, другой);
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

### Расчет данных

```python
# Добавим столбец со стоимостью в млн. руб.:
data['price_mln'] = data['last_price'] / 1000000
```

```python
# Рассчитаем цену квадратного метра и изменим тип данных:
data['price_square'] = data['last_price'] / data['total_area']
data['price_square'] = data['price_square'].astype('int')
```

```python
# Рассчитаем день недели, месяц и год публикации объявления:
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')
data['pub_weeekday'] = data['first_day_exposition'].dt.weekday
data['pub_month'] = data['first_day_exposition'].dt.month
data['pub_year'] = data['first_day_exposition'].dt.year
```

```python
# Проиндексируем этаж квартиры в зависимости от типа — первый, последний, другой:
def floor_group(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'

data['floor_group'] = data.apply(floor_group, axis=1)
data['floor_group'].value_counts()
```

```python
# Рассчитаем соотношение жилой и общей площади, а также отношение площади кухни к общей:
data['living_to_total'] = data['living_area'] / data['total_area']
data['kitchen_to_total'] = data['kitchen_area'] / data['total_area']
```

```python
data.head()
```

### Исследование полученных данных

Все данные рассчитаны, теперь изучим данные: площадь, цену, число комнат и высоту потолков, построив гистрограммы:

```python
# Построим графики по 4-м столбцам, создав функцию:
columns = ['total_area', 'price_mln', 'rooms', 'ceiling_height']

for i in columns:
    q1 = data[i].quantile(0.25)
    q3 = data[i].quantile(0.75)
    data[i].hist(bins=100, range=(q1,q3))
    plt.title(i)
    plt.show()
    print(data[i].describe())
```

### Вывод

Рассчитаны для исследований следующие параметры:

* цена квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры (варианты — первый, последний, другой);
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

Из построенных гистограмм можно сделать следующие выводы:

```python
# Выведем общее число объявлений через pivot table и отсортируем значения:
locality_top = data.pivot_table(index='locality_name', values='last_price', aggfunc='count')
locality_top = locality_top.sort_values(by='last_price', ascending=False)
locality_top.rename(columns={'last_price':'count'}, inplace=True)
```

### Изучение предложения квартир.

```python
# Выделим квартиры в городе Санкт-Петербурге:
peter_flats = data.query('locality_name == "Санкт-Петербург"')
peter_flats.head()
```

```python
# Переведем расстояние до центра в киллометры и округлим, убрав проспуки из данных:
peter_flats['city_center_km'] = peter_flats['cityCenters_nearest'] / 1000
peter_flats.dropna(subset = ['city_center_km'], inplace = True)
peter_flats['city_center_km'] = peter_flats['city_center_km'].astype('int')

# Выделим данные о расстоянии до центра и стоимость квартиры в отдельную таблицу:
peter_filter = peter_flats.pivot_table(index = 'city_center_km', values = 'price_mln')
peter_filter['price_mln'] = peter_filter['price_mln'].astype('int')
peter_filter
```

```python
# Построим график зависимости стоимости квартиры от удаленности от центра:
peter_filter.plot()
```

График сильно меняется около 7 километров, что является протяженностью исторического центра, соотвественственно это будет являться центральной зоной для сегментации квартир в центре.
